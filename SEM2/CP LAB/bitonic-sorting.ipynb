```python
def bitonic_sort(arr, up=True):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    first = bitonic_sort(arr[:mid], True)
    second = bitonic_sort(arr[mid:], False)
    return bitonic_merge(first + second, up)

def bitonic_merge(arr, up):
    if len(arr) == 1:
        return arr
    mid = len(arr) // 2
    for i in range(mid):
        if (arr[i] > arr[i + mid]) == up:
            arr[i], arr[i + mid] = arr[i + mid], arr[i]
    left = bitonic_merge(arr[:mid], up)
    right = bitonic_merge(arr[mid:], up)

    return left + right
arr = [10, 30, 11, 20, 4, 2, 1, 5]

sorted_arr = bitonic_sort(arr, True)

print("Sorted array:", sorted_arr)
```
