# Write a Python program to get the top three items in a shop.

```python
# printing without using in-built functions
x={"item1":45.5,"item2":35,"item3":41.3,"item4":55,"item5":24}
items=list(x.items())
print(items)
for i in range(len(items)):
    max=items[i][1]
    b=i
    for j in range(i+1,len(items)):
        if max<items[j][1]:
            max=items[j][1]
            b=j
    items[b],items[i]=items[i],items[b]
for k,v in items[:3]:
    print(k,":",v)
        
            
            
            
            
        
```

```python
# printing using in_built functions
x={"item1":45.5,"item2":35,"item3":41.3,"item4":55,"item5":24}
for k,v in sorted(x.items(),key=lambda item: item[1],reverse=True):  #data.items() â†’ dict_items([('a', 5), ('c', 2), ('b', 8)])
    print(k,":",v) #key=lambda item: item[1] means When sorting, use the second element (value) of each tuple as the sorting key
```

# Write a Python program to find the specified number of maximum values in a given dictionary.

```python
x={"a":5,"b":14,"c":32,"d":35,"e":24,"f":100,"g":57,"h":8,"i":100}
n=int(input("enter the number of highest values req:"))
items=list(x.items())
print(items)
for i in range(len(items)):
    max=items[i][1]
    b=i
    for j in range(i+1,len(items)):
        if max<items[j][1]:
            max=items[j][1]
            b=j
    items[b],items[i]=items[i],items[b]
for k,v in items[:n]:
    print(k)
```

# Write a Python function that evaluates mathematical expressions represented asdictionaries. The expression dictionary should support basic operations (+, -, *, /), variables,and nested expressions.

```python
def eval_expr(expr, vars={}):
    if isinstance(expr, (int, float)):  
        return expr
    if isinstance(expr, str):            
        return vars[expr]
    if "var" in expr:                    
        return vars[expr["var"]]
    
    op = expr["op"]
    args = [eval_expr(a, vars) for a in expr["args"]]

    if op == "+": return sum(args)
    if op == "-": return args[0] - sum(args[1:])
    if op == "*":
        res = 1
        for a in args: res *= a
        return res
    if op == "/":
        res = args[0]
        for a in args[1:]: res /= a
        return res
expr = {
    "op": "+",
    "args": [
        {"op": "*", "args": [2, {"var": "x"}]},
        5
    ]
}

print(eval_expr(expr, {"x": 3}))
```
