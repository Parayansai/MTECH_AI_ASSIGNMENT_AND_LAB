# Write a program that has classes such as Student, Course and Department.Enroll a student in a course of a particular department.

```python
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []     

    def add_course(self, course):
        self.courses.append(course)


class Course:
    def __init__(self, name, course_id, department):
        self.name = name
        self.course_id = course_id
        self.department = department  
        self.students = []            

        department.add_course(self)    

    def add_student(self, student):
        self.students.append(student)


class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        self.courses = []           

    def enroll(self, course):
        self.courses.append(course)
        course.add_student(self)    


if __name__ == "__main__":
    cs = Department("Computer Science")

    c1 = Course("Intro to Programming", 101, cs)
    c2 = Course("Data Structures", 201, cs)

    s1 = Student("Alice", 1)
    s2 = Student("Bob", 2)

    s1.enroll(c1)
    s2.enroll(c1)
    s1.enroll(c2)

    print("\nDepartment:", cs.name)
    for course in cs.courses:
        print(f"Course: {course.name} (ID {course.course_id})")
        print("Students:", [s.name for s in course.students])
```

# Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.

```python
class Bill:
    def __init__(self, bill_no, amount):
        self.bill_no = bill_no
        self.amount = amount

    def pay(self):
        pass


class CashPayment(Bill):
    def __init__(self, bill_no, amount):
        super().__init__(bill_no, amount)

    def pay(self):
        print(f"Bill {self.bill_no} of ₹{self.amount} paid in CASH.")


class ChequePayment(Bill):
    def __init__(self, bill_no, amount, cheque_no, bank_name):
        super().__init__(bill_no, amount)
        self.cheque_no = cheque_no
        self.bank_name = bank_name

    def pay(self):
        print(f"Bill {self.bill_no} of ₹{self.amount} paid by CHEQUE.")
        print(f"Cheque Number: {self.cheque_no}, Bank: {self.bank_name}")


b1 = CashPayment("B1001", 2500)
b2 = ChequePayment("B1002", 5200, "CHQ123456", "State Bank of India")

b1.pay()
b2.pay()
```

# Q3.Write a program to overload the - =operator to subtract two Distance Objects.

```python
class Distance:
    def __init__(self, meters):
        self.meters = meters

    def __isub__(self, other):
        if isinstance(other, Distance):
            self.meters -= other.meters
            return self
        else:
            raise TypeError("Can only subtract Distance objects")

    def display(self):
        print("Distance:", self.meters, "meters")


d1 = Distance(100)
d2 = Distance(30)

print("Before subtraction:")
d1.display()

d1 -= d2  

print("After subtraction:")
d1.display()
```

# Write a program with two classes for calculating the distance. On hasdistance specified in meters and other has distance in kilometres. Thesefunctions take argument of class Distance that converts the distance into kilometres and meters.

```python
class MeterDistance:
    def __init__(self, meters):
        self.meters = meters

    def to_kilometres(self):
        return self.meters / 1000

    def convert_from_kilometres(self, km_obj):
        if isinstance(km_obj, KilometreDistance):
            return km_obj.kilometres * 1000
        else:
            print("Invalid object passed!")

    def display(self):
        print("Distance in meters:", self.meters)


class KilometreDistance:
    def __init__(self, kilometres):
        self.kilometres = kilometres

    def to_meters(self):
        return self.kilometres * 1000

    def convert_from_meters(self, m_obj):
        if isinstance(m_obj, MeterDistance):
            return m_obj.meters / 1000
        else:
            print("Invalid object passed!")

    def display(self):
        print("Distance in kilometres:", self.kilometres)


m1 = MeterDistance(1500)
k1 = KilometreDistance(2.5)

print("Direct conversions:")
print("1500 meters =", m1.to_kilometres(), "km")
print("2.5 km =", k1.to_meters(), "meters")

print("\nConversions using object arguments:")
print("Meters from k1 =", m1.convert_from_kilometres(k1))
print("Kilometres from m1 =", k1.convert_from_meters(m1))
```
