# 1. Write a Python function find_smallest_multiple(n: int) that uses a while loop to find the smallest positive integer x such that x is divisible by all numbers from 1 to n.

```python
def find_smallest_multiple(n: int):
    if n == 1:
        return 1   # Edge case
    
    x = 1
    while True:
        divisible = True
        for i in range(1, n + 1):
            if x % i != 0:
                divisible = False
                break
        if divisible:
            return x
        x += 1
```

# 2. Write a function cubesum() that accepts an integer and returns the sum of the cubes of individual digits of that number. Use this function to make functions PrintArmstrong() and isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number.

```python
def cubesum(num):
    s = 0
    for d in str(num):
        s += int(d) ** 3
    return s
```

```python
def isArmstrong(num):
    return cubesum(num) == num
```

```python
def PrintArmstrong():
    for i in range(1, 1001):
        if isArmstrong(i):
            print(i)
```

# Write a program to input a decimal number and print its equivalent binary, octal, and hexadecimal using operators.

```python
num = int(input("Enter a decimal number: "))

# Convert to binary
n = num
binary = ""
while n > 0:
    binary = str(n % 2) + binary
    n //= 2

# Convert to octal
n = num
octal = ""
while n > 0:
    octal = str(n % 8) + octal
    n //= 8

# Convert to hexadecimal
hex_digits = "0123456789ABCDEF"
n = num
hexa = ""
while n > 0:
    hexa = hex_digits[n % 16] + hexa
    n //= 16

print("Binary:", binary)
print("Octal:", octal)
print("Hexadecimal:", hexa)
```

# Write a program that takes a sentence as input and counts the frequency of vowels, consonants, digits, and special characters separately

```python
sentence = input("Enter a sentence: ")

vowels = "aeiouAEIOU"
v = c = d = s = 0

for ch in sentence:
    if ch in vowels:
        v += 1
    elif ch.isalpha():
        c += 1
    elif ch.isdigit():
        d += 1
    else:
        s += 1

print("Vowels:", v)
print("Consonants:", c)
print("Digits:", d)
print("Special Characters:", s)
```

# Write a Python program to create a dictionary of studentsâ€™ names as keys and their marks as values.Then:A. Print the student with the highest marks ,B. Print the student with the lowest marks

```python
students = {
    "Aman": 88,
    "Riya": 92,
    "Karan": 75,
    "Sneha": 95,
    "Vikram": 81
}

# A. Highest marks
highest = max(students, key=students.get)
print("Highest Marks:", highest, students[highest])

# B. Lowest marks
lowest = min(students, key=students.get)
print("Lowest Marks:", lowest, students[lowest])
```
