# Write a Python program that takes a string and encode it that the amount of symbols would be represented by integer and the symbol.

```python
def encode_string(s):
    if not s:
        return ""
    result = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            result.append(str(count) + s[i - 1])
            count = 1
    result.append(str(count) + s[-1])
    return "".join(result)
    
print(encode_string("AAAABBBCCDAAA")) 
print(encode_string("PHP"))          
print(encode_string("hkewjhowfuhi"))
```

# Write a function that takes a mathematical equation as a string and solves for x.

```python
def solve(equation):
    left, right = equation.split("=")
    left = left.strip()
    right = right.strip()
    a = eval(left.replace("x", "1")) - eval(left.replace("x", "0"))
    b = eval(left.replace("x", "0"))
    rhs = eval(right)
    x = (rhs - b) / a
    return x
    
print(solve("2*x + 3 = 7")) 
print(solve("5*x - 8 = 22"))
```

# Write a Python program to create a new array such that each element at index i of the new array is the product of all the numbers of a given array of integers except the one at i.

```python
def product_except_self(arr):
    n = len(arr)
    result = [1]*n
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix = prefix*arr[i]
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] = result[i]*suffix
        suffix = suffix*arr[i]
    return result

print(product_except_self([10, 20, 30, 40, 50]))
print(product_except_self([1, 2, 0, 4]))
```
