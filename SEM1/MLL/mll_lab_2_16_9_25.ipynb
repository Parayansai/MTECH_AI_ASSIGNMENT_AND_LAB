```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,color="red")#df.age returns a series of 1d data of the columns 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[["age"]],df.bought_insurance,train_size=0.7)#df[["age"]] returns a 2d dataframe , if only single bracket was used 1d series of data would be returned
# the reason we want a 2d dataframe for feature data is beacause the ml algorithms expect a 2d shape as input cause the rows represent no. of datapoints and columns re[present the features and target variables expect 1d as input
```

```python
print(x_train)
print(y_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict_proba(x_test)
```
